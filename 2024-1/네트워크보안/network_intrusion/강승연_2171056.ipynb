```python
import os
import numpy as np
import pandas as pd
import glob
from google.colab import drive

drive.mount('/content/drive')

train_path = '/content/drive/My Drive/Colab Notebooks/NetworkSecurity/dataset/train_set'
test_path = '/content/drive/My Drive/Colab Notebooks/NetworkSecurity/dataset/test_set'

train_files = glob.glob(os.path.join(train_path, '*.csv'))
test_files = glob.glob(os.path.join(test_path, '*.csv'))

print(f'Train files: {train_files}')
print(f'Test files: {test_files}')

train_dataframes = []
test_dataframes = []

train_Y = []

# 열 Drop
columns_to_use = list(range(1, 13)) + [15, 17, 19]

# train file 추출 및 데이터프레임으로 변환
for file in train_files:
    df = pd.read_csv(file, usecols=columns_to_use)
    train_dataframes.append(df)
    if file.split('_')[2] == 'malicious' :
      append = 1
    else : append = 0
    for _ in range(len(df)):
        train_Y.append(append)

# test file 추출
for file in test_files:
    df = pd.read_csv(file, usecols=columns_to_use)
    test_dataframes.append(df)

# csv 파일이 여러개라 합친다.
train_x = pd.concat(train_dataframes, axis=0)
test_x = pd.concat(test_dataframes, axis=0)

del train_dataframes
del test_dataframes
```

```python
# test 파일 위에 column 명이 없어 train과 동일하게 부여
test_x.columns = train_x.columns
```

```python
# 데이터프레임 결합
train_len = len(train_x)
test_len = len(test_x)
combined_X = pd.concat([train_x, test_x], axis=0)

del train_x

# categorize

combined_X['_ws.col.Protocol'] = combined_X['_ws.col.Protocol'].fillna(0).astype('category').cat.codes
combined_X['ip.src'] = combined_X['ip.src'].fillna(0).astype('category').cat.codes
combined_X['ip.dst'] = combined_X['ip.dst'].fillna(0).astype('category').cat.codes
combined_X['http.request.method'] = combined_X['http.request.method'].fillna(0).astype('category').cat.codes
combined_X['http.connection'] = combined_X['http.connection'].fillna(0).astype('category').cat.codes
combined_X['http.content_type'] = combined_X['http.content_type'].fillna(0).astype('category').cat.codes

# 다시 원래 데이터프레임으로 분리
train_X = combined_X.iloc[:train_len, :].reset_index(drop=True)
test_X = combined_X.iloc[train_len:, :].reset_index(drop=True)

del combined_X

print("Encoded Train DataFrame:")
train_X.head()
```

```python
print(test_X.shape)
print(test_x.shape)
```

```python
import xgboost as xgb

# train_Y를 XGB 모델에 넣기 위해 Series 타입으로 변경
train_y = pd.Series(train_Y)
del train_Y

# 모델 학습 및 예측
model = xgb.XGBClassifier().fit(train_X, train_y)
test_y_pred = model.predict(test_X)

# categorize 된 버전이 아닌 원본파일으로 malicious 한 파일을 추출
indices = np.where(test_y_pred == 1)[0]
test_result = test_x.iloc[indices]

print(test_result)
```
